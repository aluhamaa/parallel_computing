```python
%matplotlib notebook
import numpy as np
import matplotlib.pyplot as plt
```

```python
%time
n_bins = 50 ## nr oh histogram bins
n_test = 10000000 ## nr of test samples
hist_max = 1. ## max value to plot on histogram
b = np.random.normal(size=n_test)**2 
```

```python
hist_max = 1
c = b[np.where(b < hist_max)[0]]
weights = np.ones_like(c) / len(c)
plt.hist(c, bins = np.arange(0, hist_max, hist_max/n_bins), weights = weights)
plt.show()
```
